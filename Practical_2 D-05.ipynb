```python
import numpy as np
import time 
import string
```

```python
training_data_file = 'eminem_songs_lyrics.txt'
```

```python
def read_training_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()
```

```python
def remove_punctuation(sentence):
    return sentence.translate(str.maketrans('', '', string.punctuation))
```

```python
def preprocess_data(data): 
    cleaned_data = remove_punctuation(data)
    cleaned_data = cleaned_data.lower()
    tokenized_data = cleaned_data.split('\n')
    
    l=[]
    for data in tokenized_data:
       if data=='':
           pass
       else:
         l.append(data.split()) 

    return l
```

```python
training_data=read_training_data(training_data_file)
```

```python
cleaned_data=preprocess_data(training_data)
```

```python
def generateText(first_word,second_word,third_dict):
    retL=[]
    fwv=list(first_word.values())
    fwk=list(first_word.keys())
    f=fwk[fwv.index(max(fwv))]
    retL.append(f)
    swv=list(second_word[f].values())
    swk=list(second_word[f].keys())
    s=swk[swv.index(max(swv))]
    
    retL.append(s)
    zwv=list(third_dict[(f,s)].values())
    zwk=list(third_dict[(f,s)].keys())
    z=zwk[zwv.index(max(zwv))]
    
    while z!="END":
        retL.append(z)
        zwv=list(third_dict[(s,z)].values())
        zwk=list(third_dict[(s,z)].keys())
        z1=zwk[zwv.index(max(zwv))]
        
        s=z
        z=z1
        
        if z=="END":
            
            return retL
            
    return retL
```

```python
sentence=generateText(first_word,second_word,third_dict)
print(sentence)
for i in sentence:
    print(i,end=" ")
```

```python
def generateProb(first_word,second_word,third_dict,sent):
    prob=first_word[sent[0]]*second_word[sent[0]][sent[1]]
    for i in range(len(sent)-1,1,-1):
        prob*=(third_dict[(sent[i-2],sent[i-1])][sent[i]])
    return prob
```
